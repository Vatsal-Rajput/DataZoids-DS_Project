```python
2.Analysis On Education Variable:
```

```python
main_df['Education'].unique()
```

```python
#CHANGING CATEGORY INTO "UG" AND "PG" ONLY....
main_df['Education'] = main_df['Education'].replace(['PhD','2n Cycle','Graduation', 'Master'],'Post Graduate')  
main_df['Education'] = main_df['Education'].replace(['Basic'], 'Under Graduate')
```

```python
#CHECKING NUMBER OF UNIQUE CATEGORIES PRESENT IN THE "Education"
print("Unique categories present in the Education:",main_df["Education"].value_counts())
print('\n')

#VISUALIZING THE "Education"
main_df['Education'].value_counts().plot(kind='bar',color = 'blue',edgecolor = "black",linewidth = 3)
plt.title("Frequency Of Each Category in the Education Variable \n",fontsize=24)
plt.figure(figsize=(8,8))
```

```python
#OBSERVATIONS:-  *97.58% of Customers in the dataset are "Post Graduate".
#                *2.410% of Customers in the dataset are "Under Graduate".
```

```python
# 3.Analysis On Marital_Status Variable:-
```

```python
main_df['Marital_Status'].unique()
```

```python
#REPLACING THE CONFLICT VALUES IN Marital_status..
main_df['Marital_Status'] = main_df['Marital_Status'].replace(['Married', 'Together'],'Relationship')
main_df['Marital_Status'] = main_df['Marital_Status'].replace(['Divorced', 'Widow', 'Alone', 'YOLO', 'Absurd'],'Single')
```

```python
#CHECKING NUMBER OF UNIQUE CATEGORIES PRESENT IN THE "Marital_Status"
print("Unique categories present in the Marital_Status:",main_df['Marital_Status'].value_counts())
print("\n")


#VISUALIZING THE "Marital_Status"
main_df['Marital_Status'].value_counts().plot(kind='bar',color = 'turquoise',edgecolor = "black",linewidth = 3)
plt.title("Frequency Of Each Category in the Marital_Status Variable \n",fontsize=24)
plt.figure(figsize=(8,8))
```

```python
#OBSERVATIONS:-  *64.46% of Customers in the dataset are in "Relationship".
#                *35.53% of Customers in the dataset are "Single".
```
